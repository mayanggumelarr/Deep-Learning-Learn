# Multiclass-Classification
Di materi ini, kita akan belajar menggunakan model neural network untuk mengklasifikasikan spesies bunga iris.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Input
from sklearn.utils.sparsefuncs import min_max_axis
```

```python
df = pd.read_csv('https://drive.google.com/uc?id=1roJ83AbgzDcvRr0Gwud0BmdUQx-oSG-w')
df.head()
```

```python
# drop ID
df = df.drop(columns='Id')
```

```python
df.head()
```

```python
# Encoding dengan one-hot encoding
category = pd.get_dummies(df.Species, dtype=int)
category
```

```python
# Gabungkang hasil encoding dengan data asli

df_new = pd.concat([df, category], axis=1)
df_new = df_new.drop(columns='Species')
df_new
```

```python
# Ubah ke numpy array
dataset = df_new.values
dataset
```

```python
# Pisahkan label dan feature independen
X = dataset[:,0:4]
y = dataset[:,4:7]
```

```python
# normalisasi
from sklearn.preprocessing import  MinMaxScaler

scalerr = MinMaxScaler()
X_scale = scalerr.fit_transform(X)
X_scale
```

```python
# splitting
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# model build
model = Sequential([
    Input(shape=(4,)),
    Dense(64, activation='relu'),
    Dense(64, activation='relu'),  #relu ubah  semua nilai negatif jadi 0, poritif tetap
    Dense(3, activation='softmax')  # 3 karena ada 3 kategori
])

model.compile(optimizer='Adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, y_train, epochs=100)
```

```python
print("untuk training")
model.evaluate(X_train, y_train, batch_size=1)

print("\nuntuk testing")
model.evaluate(X_test, y_test, batch_size=1)
```

```python
import joblib

joblib.dump(model,"iris_model.h5")
```
